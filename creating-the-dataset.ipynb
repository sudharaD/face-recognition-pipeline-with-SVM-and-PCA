````python
# Load images one by one
data = []
target = []

# creat CascadeClassifier instance
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Creating the image paths
for dir_name in folders:
    dir_path = os.path.join(data_path, dir_name)
    img_names = os.listdir(dir_path)

    for img_name in img_names:
        img_path = os.path.join(dir_path, img_name)
        img = cv2.imread(img_path)

        gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray)

        for (x, y, w, h) in faces:

            # for testing purposes can draw a rectangle around the faces
            # cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
            
            cropped_face=gray[y:y+h,x:x+w]
            cv2.imshow('cropped_face', cropped_face)
            print("If you accept this as a cropped face enter 'y' or enter 'n':")

            k = cv2.waitKey(0)

            if (k==121):
                cropped_face = cv2.resize(cropped_face, (50, 50))
                data.append(cropped_face)
                target.append(categories_dic[dir_name])

            cv2.destroyAllWindows()
````

````python
data = np.array(data)
target = np.array(target)

print(data.shape, target.shape)
````

````python
# Reshape the data array to 1D
data = data.reshape(data.shape[0], data.shape[1]*data.shape[2])

# Save the data & target files
np.save('data/data', data)
np.save('data/target', target)
````
