````python
# Import libraries
import os
import numpy as np
import pandas as pd
import cv2
````

````python
# Identify the category folders
data_path = r'data\train_data'

folders = os.listdir(data_path)
print(type(folders), type(folders[0]), folders)
````

````python
# Create the category values to numerical
categories_dic = {}
for i in range(len(folders)):
    categories_dic[folders[i]] = i

print(categories_dic)
````

````python
# Load images one by one

# Creating the image paths
for dir_name in folders:
    dir_path = os.path.join(data_path, dir_name)
    img_names = os.listdir(dir_path)

    for img_name in img_names:
        img_path = os.path.join(dir_path, img_name)
        img = cv2.imread(img_path)
        cv2.imshow('img', img)
        k = cv2.waitKey(100)
        if k == 27:
            break

cv2.destroyAllWindows()
````
